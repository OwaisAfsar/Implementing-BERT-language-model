```python
import pandas as pd
```

```python
train_df = pd.read_csv('olid-training-v1.0.csv', header=None)
train_df.head()
```

```python
test_df = pd.read_csv('testset-levela.csv', header=None)
test_df.head()
```

```python
train_df_bert = pd.DataFrame({
    '0':range(len(train_df)),
    '2':train_df[2],
    'alpha':['a']*train_df.shape[0],
    '1': train_df[1].replace(r'\n', ' ', regex=True)
})

train_df_bert.head()
```

```python
dev_df_bert = pd.DataFrame({
    '0':range(len(test_df)),
    'alpha':['a']*test_df.shape[0],
    '1': test_df[1].replace(r'\n', ' ', regex=True)
})

dev_df_bert.head()
```

```python
train_df_bert.to_csv('train.tsv', sep='\t', index=False, header=False)
```

```python
dev_df_bert.to_csv('dev.tsv', sep='\t', index=False, header=False)
```
